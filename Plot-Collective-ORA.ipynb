```python
import os, sys
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
```

```python
data_directory = "/Users/swarnavo/CodeX/CoRe/Examples/collective_action"
os.chdir(data_directory)

datas = {}

datas[0] = pd.read_csv('collectively_activated_GOBPs0.1.csv')
datas[1] = pd.read_csv('collectively_activated_GOBPs0.1-baricitinib.csv')
datas[2] = pd.read_csv('collectively_activated_GOBPs0.1-IFITM3.csv')
#datas[3] = pd.read_csv('collectively_activated_GOBPs0.1-PSMB8.csv')
```

```python
size = len(datas.keys())

all_ora = np.zeros(shape=(len(datas[0].index),size))
```

```python
for k in datas.keys():
    all_ora[:,k] = datas[k]['q_values'].to_numpy() 
```

```python
all_gos = datas[0]['go_names'].to_list()
```

```python
x_labels = ['SARS-CoV-2','SARS-CoV-2 + baricitinib','SARS-CoV-2 + IFITM3']#,'PSMB8']
```

```python
fig, ax = plt.subplots(figsize=(4,20))

plt.imshow(-np.log10(all_ora),cmap=cm.Reds,aspect='auto',vmin=0,vmax=10)

cbar = plt.colorbar(fraction=0.1,pad=0.05)

ax.set_xticks(np.arange(0,len(x_labels)))
ax.set_xticklabels(x_labels,fontsize=20,rotation=90)

ax.set_yticks(np.arange(0,len(all_gos)))
ax.set_yticklabels(all_gos,fontsize=18)

ax.set_xticks(np.arange(-.5, len(x_labels), 1), minor=True)
ax.set_yticks(np.arange(-.5, len(all_gos), 1), minor=True)

cbar.set_label(r'$-\log_{10}$($Q$)',fontsize=18,rotation=-90,labelpad=20)
cbar.ax.tick_params(labelsize=18)

print(len(all_gos))

ax.grid(which='minor', color='black', linestyle='-', linewidth=1.5, alpha=0.5)
```

```python
fig, ax = plt.subplots(figsize=(4,25))

plt.imshow(-np.log10(all_ora[50:,:]),cmap=cm.Reds,aspect='auto',vmin=0,vmax=10)

cbar = plt.colorbar(fraction=0.1,pad=0.05)

ax.set_xticks(np.arange(0,len(x_labels)))
ax.set_xticklabels(x_labels,fontsize=20,rotation=90)

ax.set_yticks(np.arange(0,len(all_gos[50:])))
ax.set_yticklabels(all_gos[50:],fontsize=18)

ax.set_xticks(np.arange(-.5, len(x_labels), 1), minor=True)
ax.set_yticks(np.arange(-.5, len(all_gos[50:]), 1), minor=True)

cbar.set_label(r'$-\log_{10}$($Q$)',fontsize=18,rotation=-90,labelpad=20)
cbar.ax.tick_params(labelsize=18)

print(len(all_gos[50:]))

ax.grid(which='minor', color='black', linestyle='-', linewidth=1.5, alpha=0.5)
```
